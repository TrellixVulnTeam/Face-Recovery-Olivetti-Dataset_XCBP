```python
# inner product of two vectors.
def innerProduct(X, Y):
    return np.inner(X, Y)
```

# 3 Breast Cancer Data

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.metrics import f1_score
```

###### A. Load the breast cancer data with the load_breast_cancer method from the module sklearn.datasets.

```python
cancer_dataset = load_breast_cancer()
```

```python
# Extract features and the target from the above dataset
X, y = load_breast_cancer(return_X_y = True) # X is the features, and y is the target
```

###### B. Standardize each feature in the data set.

```python
# feature scaling and standardization
scaler = StandardScaler() 
# standardization of the features through the StandardScaler
X_sc = scaler.fit_transform(X)
```

###### D. Perform k-means with k = 2 on the full set of features, and on the first 2 principle components only. Compare how well the clusters found by k-means in each of these cases compare to the true targets of the data set.

```python
# KMeans Model on All Features
kmeans1 = KMeans(n_clusters=2) # use k = 2 as n_clusters (number of clusters)  

# perform k-means on the full set of features 
kmeans1.fit(X_sc)  # (X_sc): the standardized features

labels1 = kmeans1.labels_

# KMeans Model on the first 2 PCA Components
kmeans2 = KMeans(n_clusters=2) # use k = 2 as n_clusters (number of clusters)  

# perform k-means on the first 2 principal components 
kmeans2.fit(pca_X)  # (pca_X): the standardized features

labels2 = kmeans2.labels_
```

```python

# subplot container
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('K-Means Clustering')
fig.set_figwidth(14)

# first plot KMeans Model on All Features
ax1.scatter(X_sc[:,0], X_sc[:,1], c=labels1, cmap=plt.cm.rainbow)
ax1.set_xlabel('Standardized feature 1')
ax1.set_ylabel('Standardized feature 2')
ax1.set_title("K-Means on the full set of features")

# second plot KMeans Model on PCA Components 
ax2.scatter(pca_X[:,0], pca_X[:,1], c=labels2, cmap=plt.cm.rainbow)
ax2.set_xlabel('Standardized feature 1')
ax2.set_ylabel('Standardized feature 2')
ax2.set_title("K-Means on the first 2 Principal Components")

plt.show()
```

```python
# Evaluating the K-Means on All Features
f1_score(labels1, kmeans1.predict(X), average='weighted', labels=np.unique(kmeans1.predict(X)))
```

```python
# Evaluating the Model on PCA Components
f1_score(labels2, kmeans2.predict(pca_X), average='weighted', labels=np.unique(kmeans2.predict(pca_X)))
```

The K-Means model on all features has an f1_score of about 49%, which is halfway accurate compared to the true targets of the dataset. On the other hand, the K-Means model on standardized two PCA components has an f1_score of 100%, which implies it is a true reflection of the targets as depicted in the data set. It's plot also shows a clear distinction between the clusters compared to the plot of the model acting on all features. 

# 4 Olivetti Faces

```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
%matplotlib inline
```

###### A. Fetch and load the data with the fetch_olivetti_faces method from the module sklearn.datasets.

```python
faces_dataset = fetch_olivetti_faces()
```

###### B. Demean each face in the data set (no need to divide by standard deviation as every dimension is a number between a fixed range representing a pixel).

```python
# get the faces in the data
faces = faces_dataset.images 
single_mean = sum(faces) / len(faces)
# subtracting the sample mean from each observation
mean = faces/len(faces)
faces = faces - mean
```

###### C. Compute and display the first 9 eigenfaces.

```python
X = faces.reshape((faces.shape[0], faces.shape[1] * faces.shape[2]))
print(X.shape)

# Use PCA as SVD in creating the matrix for eigenfaces 
n_components = 9 # the number of eigen faces needed

pca=PCA(n_components=n_components, whiten=True, svd_solver='auto')
pca.fit(X)

number_of_eigenfaces = len(pca.components_)
eigen_faces = pca.components_.reshape((number_of_eigenfaces, faces.shape[1], faces.shape[2]))

cols=9 # 9 columns each for one face
rows=1 # i row of faces

# plotting to show the eigenfaces
fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(15,3))
axarr = axarr.flatten()
for i in range(number_of_eigenfaces):
    ax[i].imshow(eigen_faces[i],cmap="gray")
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title("eigenface:{}".format(i))

plt.suptitle("First 9 Eigen Faces")
```

###### D. In class we showed that any given face in the data set can be represented as a linear combination of the eigenfaces. For any face in the data set, show how it progresses as we combine 1, 51, 101, . . . eigenfaces, until the full image is recovered.

```python
k = [1]+[i + 50 for i in range(1, 400, 50)]
k
```

```python
# let x be the image
# x_coeficeints = eigen_face[i].(x - mean) , k times for k coeeficients

coefficients = []
k = 9  # number of coefficients
for i in range(k):
    coefficients.append(eigen_faces[i].dot(faces[0]))
    
print(len(coefficients))
```

```python
# recovering the image, faces [0]
# x_recovered = mean + (coefficients[i] * eigen_faces[i]) for i = 1 to i = k

# compute the sums
sums = 0
k = 9
for i in range(k):
    sums += (coefficients[i] * eigen_faces[i])

recovered_face = single_mean + sums
```

```python
# plotting the recovered face
fig,ax=plt.subplots(1,1,figsize=(6,6))
ax.imshow(recovered_face.reshape(64,64), cmap="gray")
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Recovered Face using 9 eigen_faces')
```
